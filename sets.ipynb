```python
#Sets in Python:

#Syntax:
        # My_set={1,2,3,4} -In curly braces
        #1. Mutable
        #2. Unordered
        #3. Cannot assign duplicates values
        #4. Hetrogenous- Semi It can store data types like string,number,tuple but not everything.

#How set stores value in python: It stores values in hashing data structure.

b=hash("Hello")
print(b)
```

```python
a={1,2,9,3,"Hello",4,5,6}

for i in a:
    print(i)
```

```python
#Set Methods:

s={1,2,3}

s.add(4)
s.remove(2)  #Remove and discard both are same function
print(s)

popped_element=s.pop()
print(popped_element)

s.clear()
print(s)
```

```python
a={1,2,3,4,5}
b={2,3,4,5,6}

union_set=a.union(b)
print(union_set)

print(a|b) # For union of two sets

intersection_set=a.intersection(b)
print(intersection_set)

print(a&b) # For intersection of two sets

difference_set=a.difference(b)
print(difference_set)

print(a-b)

d_set_b=b.difference(a)
print(d_set_b)
print(b-a)

symmetric_diff= a.symmetric_difference(b)  # Common part htana
print(symmetric_diff)

s_diff=a^b  #For symmetric diff power operators
print(s_diff)
```
