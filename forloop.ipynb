```python
#2.Print natural number upto n using for loop
n=int(input("Enter number throughh which you want to print it"))
for i in range(1,n+1):
    print(i)
```

```python
#3.Reverse for loop Print n to 1
n=int(input("Enter number from where to print till 1"))
for i in range(n,0,-1):
    print(i)
```

```python
#5.Sum upto n terms:

n=int(input("Enter a number till you want to sum"))
sum=0
for i in range(n+1):
    sum+=i
print(f"Your sum is {sum}")
```

```python
#6.Factorial of a number:
n=int(input("Enter a number to calculate factorial of a number:"))
fact=1
for i in range(1,n+1):
    fact=fact*i
print(f"Factorial of a given number is {fact}")
```

```python
#7.Print the suum of all even & odd numbers in a range separately.
n=int(input("Till your number"))
even=0
odd=0
for i in range(1,n+1):
    if i%2==0:
        even=even+i
    else:
        odd=odd+i

print(f"Your even and odd sum are {even},{odd} respectively")
```

```python
#8. Print the factors of the number:
n=int(input("Number that you want to find the factors are :"))

for i in range(1,n+1):
    if n%i==0:
        print(f"Factors of the {n} is {i}")
```

```python
#9. Accept a number and check it is a perfect number or not:
n=int(input("Check whether the number is a perfect number or not"))
per_sum=0
for i in range(1,n):
    if n%i==0:
        per_sum=per_sum+i

if per_sum==n:
    print(f"Perfect number is {n}")
else:
    print("Not perfect number")
```

```python
#10.Check prime or not
n=int(input("Enter number to check whether it is prime or not"))
counter=0
for i in range(1,n+1):
    if n%i==0:
        counter=counter+1
if counter==2:
    print(f"{n} is prime number")
else:
    print("Not a prime number")
```

```python
#12.Pallindrome String or not:

clg="NAMAN"
b=""
for i in range(len(clg)-1,-1,-1):
    b=b+clg[i]
if b==clg:
    print("Pallindrome")
else:
    print("Not pallindrome")
```

```python
#13.Count all letters, digits, and special symbols from a given string
#Given: str1 = "P@#yn26at^&i5ve"

str1 = "P@#yn26at^&i5ve"
letter=0
digits=0
special=0
for i in str1:
    if i.isalpha():
        letter=letter+1
    elif i.isdigit():
        digits=digits+1
    else:
        special=special+1

print(f"Your no. of letters are {letter}")
print(f"Your no. of digits are {digits}")
print(f"Your no. of special characters are {special}")
```
