```python
#Functions in Python:

#Function is a block of code that execute when we want.
#It is a reusable code we execute it multiple times.

#Parameter- The thing you accept is parameter.

#Argument- The thing you provide to parameter is argument

```

```python
def hello():
    print("This is a hello function")

hello()
```

```python
#Positional Arguments:

def sum(a,b):
    print(f"The sum of your number is :{a+b}")

sum(12,36.5)
sum(12,12)
```

```python
#Keyword Argument:

def hello(name,age):
    print(f"Your name is {name} and your age is {age}")

hello("akarsh",22)
hello(name="Saurabh",age=21)
hello(age=21,name="ARYA")
```

```python
#Default Argument:

def sum(a,b=0):
    print(f"Sum of two number is {a+b}")

sum(4,5)
sum(4)
```

```python
#Chcek strings is pallidrome or not:

def pallindrome(st):
    rev=""
    for i in range(len(st)-1,-1,-1):
        rev=rev+st[i]
    
    if rev==st:
        print("Pallindrome")
    else:
        print('Not a Pallindrome')

pallindrome("NAMAN")
pallindrome("Cursor")
```
