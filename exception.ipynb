```python
#Exception Handling:

#Exceptions:
#Exceptions are unexpected events or errors that occurs during the execution of a program.

#Which disrupts the normal flow of the program.
```

```python
a=int(input('Enter a number'))

try:
    print(10/a)

except ZeroDivisionError:
    print("Sorry you cannot divide by 0")

print('I have done division')
```
