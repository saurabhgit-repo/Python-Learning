```python
d={10:100,20:200,30:300}
print(d)

print(d[10])  #Keys act like index values

d[10]=500
print(d[10])

d.update({50:500})  # to add key value pairs
#OR
d[60]=600 #Creating new key value pairs
print(d)
  
del d[30]  #To delete 
print(d)
```

```python
#Traversing in Dictionary
d={10:100,20:200,30:300,40:400}

for i in d:
    print(i) #It return keys
    print(d[i]) #It return both key and values

for i in d.values():  #Also i can do so
    print(i)
```

```python
a=[1,2,3,4,5]
b=a

b[0]=100

print(a)  # It is also knows as deep copy
```

```python
a=[1,2,3,4,5]

b=a.copy()
b[0]=100

print(a)  #It is also knows as shallow copy
```

```python
d={10:100,20:200,30:300}
print(d.items())  #It will return key value pairs

```

```python
#Dictionary question to merge two dictinary
d1={"name":"Saurabh","age":21}
d2={"college":"MIT MEERUT","Course":"B.tech"}

for i in d2:
    d1[i]=d2[i]

print(d1)
```

```python
#Write a program to sum all the values in dictionary

d={10:100,20:300,30:300}
sum=0
for i in d:
    sum=sum+d[i]

print(sum)


```

```python
# count the frequency of each element in lst

a=[1,1,1,1,12,2,2,3,3,3,3,3,4,5]
count=0

for i in a:
    if i==1:
        count=count+1

print(count)
```

```python
# count the frequency of each element in lst

a=[1,1,1,1,12,2,2,3,3,3,3,3,4,5]
d={}
for i in a:
    if i in d.keys():
        d[i]+=1
    else:
        d[i]=1
print(d)
```

```python
#WAP to combine two dictionary by adding values for common keys.

d1={10:100,20:200,30:300}
d2={30:400,40:500,50:600}

for i in d2:
    if i in d1.keys():
        d1[i]+= d2[i]
    else:
        d1[i]=d2[i]

print(d2)

```
